```python
import numpy as np
import unittest

def solve(arr, guard_index=None, direction=(-1, 0), visited=None, count=0, vis_dir=set()):
    if guard_index is None:
        guard_index = tuple(np.argwhere(arr == "^")[0])
        visited = set()
        vis_dir = set()
        count = 0

    if guard_index not in visited:
        visited.add(guard_index)
        count += 1

    if (guard_index, direction) in vis_dir:
        return -1
    vis_dir.add((guard_index, direction))

    next_index = (guard_index[0] + direction[0], guard_index[1] + direction[1])

    if next_index[0] < 0 or next_index[1] < 0 or \
            next_index[0] >= arr.shape[0] or next_index[1] >= arr.shape[1]:
        return count

    cell = arr[next_index]

    if cell == "#":
        if direction == (-1, 0):
            direction = (0, 1)
        elif direction == (0, 1):
            direction = (1, 0)
        elif direction == (1, 0):
            direction = (0, -1)
        elif direction == (0, -1):
            direction = (-1, 0)
    else:
        guard_index = next_index

    return solve(arr, guard_index, direction, visited, count, vis_dir)


def solve_loops(arr):
    true_count = 0
    for i in range(arr.shape[0]):
        for j in range(arr.shape[1]):
            if arr[i, j] != "^":
                modified_arr = arr.copy()
                modified_arr[i, j] = "#"
                if solve(modified_arr) == -1:
                    true_count += 1
    return true_count


class TestSolve(unittest.TestCase):

    def setUp(self):
        self.test_case = np.array(
            [
                list("....#....."),
                list(".........#"),
                list(".........."),
                list("..#......."),
                list(".......#.."),
                list(".........."),
                list(".#..^....."),
                list("........#."),
                list("#........."),
                list("......#...")
            ]
        )

    def test_solve(self):
        self.assertEqual(solve(self.test_case), 41)

    def test_solve_loops(self):
        self.assertEqual(solve_loops(self.test_case), 6)

unittest.main(argv=[""], exit=False)

print(solve(read_file("day6.txt")))
print(solve_loops(read_file("day6.txt")))
```
