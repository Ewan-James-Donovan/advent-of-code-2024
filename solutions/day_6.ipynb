```python
import numpy as np
import unittest
import sys

sys.setrecursionlimit(999999) # Please don't judge me ...

def read_file(file_path):
    with open(file_path, "r") as file:
        lines = file.readlines()
        return np.array([list(line.strip()) for line in lines])

def solve(arr, guard_index=None, direction=(-1, 0), visited = None, count = 0, visited_with_direction = None, skip_loop_check = False):
    if guard_index is None:
        guard_index = tuple(np.argwhere(arr == "^")[0])
        visited = set()
        visited_with_direction = set()
        count = 0

    if guard_index not in visited:
        visited.add(guard_index)
        count += 1

    if (guard_index, direction) in visited_with_direction:
        return "loop"
    visited_with_direction.add((guard_index, direction))

    next_index = (guard_index[0] + direction[0], guard_index[1] + direction[1])
    

    if next_index[0] < 0 or next_index[1] < 0 or next_index[0] >= arr.shape[0] or next_index[1] >= arr.shape[1]:
        if not skip_loop_check:
            loop_count = 0
            for index in visited:
                if arr[index] != "^":
                    original_value = arr[index]
                    arr[index] = "#"
                    if solve(arr, skip_loop_check = True) == "loop":
                        loop_count += 1
                    arr[index] = original_value
            return [count, loop_count]
        return count

    cell = arr[next_index]

    if cell == "#":
        if direction == (-1, 0):
            direction = (0, 1)
        elif direction == (0, 1):
            direction = (1, 0)
        elif direction == (1, 0):
            direction = (0, -1)
        elif direction == (0, -1):
            direction = (-1, 0)
    else:
        guard_index = next_index

    return solve(arr, guard_index, direction, visited, count, visited_with_direction, skip_loop_check)

class TestSolve(unittest.TestCase):

    def setUp(self):
        self.test_case = np.array(
            [
                list("....#....."),
                list(".........#"),
                list(".........."),
                list("..#......."),
                list(".......#.."),
                list(".........."),
                list(".#..^....."),
                list("........#."),
                list("#........."),
                list("......#...")
            ]
        )

    def test_solve(self):
        self.assertEqual(solve(self.test_case)[0], 41)

    def test_solve_loops(self):
        self.assertEqual(solve(self.test_case)[1], 6)

unittest.main(argv=[""], exit=False)

print(solve(read_file("day6.txt")))
```
